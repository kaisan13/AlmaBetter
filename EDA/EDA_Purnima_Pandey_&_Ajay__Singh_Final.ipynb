### **Variables Description**

**1) Year:** The dataset includes the years spanning from **1970 to 2017,**documenting the time frame of the incidents.

**2) Month:** This field indicates the numerical representation of the month in which each incident took place.

**3) Day:** Numeric values in this field signify the specific day of the month when each incident occurred.

**4) Country:**This identifies the country or location where the attack occurred. In the case where the country in which an incident occurred cannot be identified, it is coded as “Unknown”.


**5) State:** This variable records the administrative region's name at the time of the incident, providing further geographical context to the events.


**6) Region:** This attribute specifies the incident's geographical area, classified into **12** distinct categories such as **North America, Central America & Caribbean, South America, East Asia, Southeast Asia, South Asia, Central Asia, Western Europe, Eastern Europe, Middle East & North Africa, Sub-Saharan Africa, and Australasia & Oceania.**


**7) City:** Holds the location name where the incident took place, including city, village, or town.

**8) Latitude:** Records the geographic latitude of the incident's city location.

**9) Longitude:** Records the geographic longitude corresponding to the incident's city.

**10) Attack Type:** This field categorizes tactics used, with nine categories including **Assassination, Hijacking, Kidnapping, Barricade Incident, Bombing/Explosion, Armed Assault, Unarmed Assault, Facility/Infrastructure Attack, and Unknown for unrecorded cases.**

**11) Killed:** This field provides data on the number of fatalities resulting from the incident, indicating the loss of life.

**12)Wounded:** This attribute records the count of individuals who sustained injuries during the occurrence of the attack, reflecting the number of people harmed.

**13) Target:** This field encompasses data about both the primary intended targets of the attack and any bystanders who were incidentally affected, providing insights into the scope of impact.

**14) Group:** This attribute holds the name of the specific group responsible for carrying out the attack, shedding light on the entity behind the incident.

**15) Weapon Type:** This category records the general classification of weapons used in the attack, offering insights into the method and severity of the incident. I


**16) Weapon Detail:**This field notes any pertinent information on the type of weapons used in the incident. Such notes could include the novel use or means of concealing a weapon, specific weapon models, interesting details of the weapons’ origins, etc.


**17) Target Type:** This attribute categorizes the general type of victim or target. It accounts for instances where individuals are targeted due to their association with specific individuals, like prominent figures. Categories encompass **government, police, military, airports and aircraft, educational institutions, food or water supply, journalists and media, maritime, non-governmental organizations, private citizens and property, religious figures and institutions, telecommunication, tourists, and violent political parties.** This classification provides a comprehensive overview of the diverse targets involved in incidents.

**18) Sub Target:** This field further refines target categories by offering more specific designations within each target type. It provides additional granularity to the target classification, enhancing the understanding of the diverse range of targets involved in incidents. If a specific subcategory doesn't apply, this field remains empty.

**19) Summary:** This field serves as a comprehensive repository of information, encapsulating details about the attack, its motive, casualties, the weapon utilized, and the location where the incident took place. It provides a concise overview of key incident attributes.

**20) Nationality:** This represents the nationality of the target that was attacked. It might not always match the country where the incident happened, although usually it does. In cases of hijackings, the nationality of the airplane is noted, not the passengers'.


**21) Motive:** This field offers an explanation for why a certain location was chosen for an attack or protest. It helps us understand the motives behind these actions, providing insights into the reasons that led to these decisions.


**22) Success:** This variable indicates whether a terrorist attack was successful or not. The description could mention the success rate based on the mean value and highlight the proportion of successful attacks.

**23) Suicide:** This variable shows whether a suicide attempt was successful or not, using a code. If the code is "1" it means the incident was a suicide attack, where the attacker intentionally ended their own life. If the code is "0" there's no clear evidence that the incident was a suicide attack.

**24) Multiple:** This category pertains to the connection between multiple attacks. When the code is "1" it indicates that there were numerous incidents that are linked together. In contrast, a code of "0" suggests that only a single incident has occurred without any connection to other events.

````python
#2- How has the frequency of terrorist attacks changed over the past decade?

# Creating a copy of the DataFrame to avoid modification warnings
gtd_df_copy = gtd_df.copy()
# Create a new column 'Decade' by dividing the 'Year' by 10 and rounding down
gtd_df_copy['Decade'] = (gtd_df_copy['Year'] // 10)*10
# Grouping the data by 'Decade' and counting the number of incidents for each decade
atkBydecade = gtd_df_copy.groupby('Decade')['Decade'].count()
# Creating a custom DataFrame
atkBydecade_df = pd.DataFrame({'Decade': atkBydecade.index, 'IncidentCount':atkBydecade.values })
atkBydecade_df
````

````python
#Q3-Are there any significant seasonal patterns in terrorist incidents?

# Grouping the data by 'Month' and counting the number of incidents for each month
incidents_per_month = gtd_df_copy.groupby('Month')['Month'].count()
# Creating a custom DataFrame
seasonal_pattern_df = pd.DataFrame({'Month': incidents_per_month.index, 'IncidentCount': incidents_per_month.values})
# Sorting the DataFrame by month
seasonal_pattern_df = seasonal_pattern_df.sort_values('Month')
seasonal_pattern_df = seasonal_pattern_df[seasonal_pattern_df['Month'] != 0]
# Display the custom DataFrame

season_df = {'Spring': 0, 'Summer':0, 'Winter':0}
summer = [3, 4, 5, 6]
spring = [7,8,9,10]
winter = [11, 12, 1, 2]
for month in range(1, 13):
  if month in spring:
    season_df['Spring'] += seasonal_pattern_df.loc[seasonal_pattern_df['Month'] == month, 'IncidentCount'].values[0]
  elif month in summer:
    season_df['Summer'] += seasonal_pattern_df.loc[seasonal_pattern_df['Month'] == month, 'IncidentCount'].values[0]
  elif month in winter:
    season_df['Winter'] += seasonal_pattern_df.loc[seasonal_pattern_df['Month'] == month, 'IncidentCount'].values[0]
# Create a DataFrame from the dictionary
season_df = pd.DataFrame(list(season_df.items()), columns=['Season', 'IncidentCount'])
season_df
````

````python
#4- What are the primary targets of terrorist attacks, and have they shifted over time?
TargetCount_df = gtd_df_copy.groupby(['Year','Target_type'])['Target_type'].count().unstack()
TargetCount_df = TargetCount_df.drop(columns = ['Unknown'], errors = 'ignore')
TargetCount_df = TargetCount_df.fillna(0.0)
TargetCount_df.head(10)
````

````python
#Q5-How does the number of terrorist incidents vary between developed and developing countries?
country_development = {}
developed_countires = ['United States', 'Canada', 'United Kingdom', 'Germany', 'France', 'Japan', 'Australia', 'South Korea', 'New Zealand', 'Singapore', 'Norway', 'Sweden', 'Denmark', 'Switzerland', 'Netherlands', 'Finland', 'Belgium', 'Austria', 'Ireland', 'Iceland', 'Luxembourg', 'Italy', 'Spain']
developing_countries = ['Afghanistan', 'Bangladesh', 'Cambodia', 'Democratic Republic of Congo', 'Ethiopia', 'Haiti', 'India', 'Kenya', 'Nigeria', 'Pakistan', 'Philippines Sudan', 'Tanzania', 'Uganda', 'Yemen', 'Zambia', 'Zimbabwe', 'Bolivia', 'Guatemala', 'Honduras', 'Myanmar', 'Nepal', 'Nicaragua', 'Rwanda', 'Senegal']
# gtd_df_copy['Country'].unique()
for country in gtd_df_copy['Country']:
    if country in developed_countires:
      country_development[country] = 'Developed'
    elif country in developing_countries:
      country_development[country] = 'Developing'
    else:
      country_development[country] = 'Under-developed'
# Create a new column 'CountryType' based on the mapping
gtd_df_copy['CountryType'] = gtd_df_copy['Country'].map(country_development)
# Group data by countrytype and count number of incidents
incidents_country_df = gtd_df_copy.groupby('CountryType')['CountryType'].count()
incidents_country_df = pd.DataFrame({"Country_type":incidents_country_df.index,"Counts":incidents_country_df.values})
 #Convert Counts into Percentage
total_counts = incidents_country_df['Counts'].sum()
incidents_country_df['Percent'] = round((incidents_country_df['Counts']/total_counts)*100, 2)
incidents_country_df
````
